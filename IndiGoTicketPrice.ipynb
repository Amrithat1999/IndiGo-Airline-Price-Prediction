# Data Cleaning

```python
data.drop(['Additional_Info','Airline'],axis=1,inplace=True)
```

```python
#replace Stopages

data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
# lets create dummy variables of these categorical columns to convert them into numerical columns

dummy=pd.get_dummies(data[['Source','Destination']])
dummy
```

```python
# lets check the head of the dataset

data.head()
```

```python
#Departure Date

data["Journey_day"] = pd.to_datetime(data.Dep_Date, format="%d-%m-%Y").dt.day
data["Journey_month"] = pd.to_datetime(data["Dep_Date"], format = "%d-%m-%Y").dt.month
```

```python
# Extracting Minutes& hour

data["Dep_hour"] = pd.to_datetime(data["Dep_Time"]).dt.hour
data["Dep_min"] = pd.to_datetime(data["Dep_Time"]).dt.minute
```

```python
#Arrival Date

data["Arrival_day"] = pd.to_datetime(data.Arrival_Date, format="%d-%m-%Y").dt.day
data["Arrival_month"] = pd.to_datetime(data["Arrival_Date"], format = "%d-%m-%Y").dt.month
```

```python
# Extracting Minutes& hour

data["Arrival_hour"] = pd.to_datetime(data.Arrival_Time).dt.hour
data["Arrival_min"] = pd.to_datetime(data.Arrival_Time).dt.minute
```

```python
data=pd.concat([data,dummy],axis=1)
data.head()
```

```python
data.shape
```

```python
data.drop(['Source','Destination','Dep_Date','Dep_Time','Arrival_Date','Arrival_Time'],axis=1,inplace=True)
```

```python
data.shape
```

```python
# columns in x Data

data.columns
```

```python
data.head()
```

```python
x=data.drop(['Price'],axis=1)
y=data.Price
```

```python
#split data

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
# columns in x Data

x.columns
```

```python
x.shape
```

# Create Model

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
```

```python
model.fit(xtrain,ytrain)
```

```python
y_pred=model.predict(xtest)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score,mean_absolute_error
from math import sqrt
```

```python
evs=explained_variance_score(ytest,y_pred)
print('EVS:',evs)
mae=mean_absolute_error(ytest,y_pred)
print('MAE:',mae)
mse=mean_squared_error(ytest,y_pred)
print('MSE:',mse)
rmse=sqrt(mse)
print('RMSE:',rmse)
```

```python
from sklearn import metrics

metrics.r2_score(ytest,y_pred)
```
